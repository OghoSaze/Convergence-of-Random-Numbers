```python
# Set the number of samples and the number of terms in the partial sum
num_samples = 10000
num_terms = 100
```

```python
# Define the random variables
def Y1(X):
    return X

def Y2(X):
    return X**(-1/4)

def Y3(X):
    return X**(-2/3)

def Y4(X):
    return np.tan(np.pi * (X - 1/2))
```

```python
# Function to compute the partial sum Dn(Y)
def partial_sum(Y, n):
    return np.sum(Y[:n])
```

```python
# Generate samples from the uniform distribution U(0, 1)
X = np.random.uniform(0, 1, num_samples)
```

```python
# Compute the partial sums for each random variable
partial_sums_1 = np.zeros(num_terms)
partial_sums_2 = np.zeros(num_terms)
partial_sums_3 = np.zeros(num_terms)
partial_sums_4 = np.zeros(num_terms)
```

```python
for n in range(num_terms):
    partial_sums_1[n] = partial_sum(Y1(X), n+1)
    partial_sums_2[n] = partial_sum(Y2(X), n+1)
    partial_sums_3[n] = partial_sum(Y3(X), n+1)
    partial_sums_4[n] = partial_sum(Y4(X), n+1)
```

```python
# Plot the realizations of the sequence (Dn(Y))
plt.plot(range(1, num_terms+1), partial_sums_1, label='Y1')
plt.plot(range(1, num_terms+1), partial_sums_2, label='Y2')
plt.plot(range(1, num_terms+1), partial_sums_3, label='Y3')
plt.plot(range(1, num_terms+1), partial_sums_4, label='Y4')
plt.xlabel('Number of terms (n)')
plt.ylabel('Partial sum (Dn(Y))')
plt.title('Convergence of the sequence (Dn(Y))')
plt.legend()
plt.show()
```

```python
# Function to compute the partial sum Dn(Y)
def partial_sum(Y, n):
    return np.sum(Y[:n])
```

```python
for n in range(num_terms):
    partial_sums_1[n] = partial_sum(Y1(X), n+1)
    partial_sums_2[n] = partial_sum(Y2(X), n+1)

# Compute the expectations E[Y]
E_Y1 = np.mean(Y1(X))
E_Y2 = np.mean(Y2(X))

# Compute the standard deviations σ(Y)
sigma_Y1 = np.std(Y1(X))
sigma_Y2 = np.std(Y2(X))

# Compute the asymptotic barrier ±σ(Y)
asymptotic_barrier_1 = np.sqrt(2 * np.log(np.log(np.arange(1, num_terms+1)))) * sigma_Y1
asymptotic_barrier_2 = np.sqrt(2 * np.log(np.log(np.arange(1, num_terms+1)))) * sigma_Y2

# Compute the scaled difference √n · (Dn(Y) - E[Y]) / (√(2 log(log(n))))
# Compute the scaled difference √n · (Dn(Y) - E[Y]) / (√(2 log(log(n))))
scaled_diff_1 = np.sqrt(np.arange(1, num_terms+1)) * (partial_sums_1 - E_Y1) / (np.sqrt(2 * np.log(np.log(np.arange(1, num_terms+1)))))
scaled_diff_2 = np.sqrt(np.arange(1, num_terms+1)) * (partial_sums_2 - E_Y2) / (np.sqrt(2 * np.log(np.log(np.arange(1, num_terms+1)))))
```

```python
# Plot the evolution of √n · (Dn(Y) - E[Y]) / (√(2 log(log(n)))) and the asymptotic barrier
plt.plot(range(1, num_terms+1), scaled_diff_1, label='Y1')
plt.plot(range(1, num_terms+1), asymptotic_barrier_1, 'r--', label='Asymptotic Barrier')

plt.xlabel('Number of terms (n)')
plt.ylabel('Scaled Difference')
plt.title('Evolution of √n · (Dn(Y) - E[Y]) / (√(2 log(log(n)))) for Y1')
plt.legend()
plt.show()

plt.plot(range(1, num_terms+1), scaled_diff_2, label='Y2')
plt.plot(range(1, num_terms+1), asymptotic_barrier_2, 'r--', label='Asymptotic Barrier')

plt.xlabel('Number of terms (n)')
plt.ylabel('Scaled Difference')
plt.title('Evolution of √n · (Dn(Y) - E[Y]) / (√(2 log(log(n)))) for Y2')
plt.legend()
plt.show()
```

```python
num_samples = 10000
num_terms = 100

# Define the random variable
def Y3(X):
    return X**(-2/3)

# Function to compute the partial sum Dn(Y)
def partial_sum(Y, n):
    return np.sum(Y[:n])

# Generate samples from the uniform distribution U(0, 1)
X = np.random.uniform(0, 1, num_samples)

# Compute the partial sums for the random variable Y3
partial_sums_3 = np.zeros(num_terms)

for n in range(num_terms):
    partial_sums_3[n] = partial_sum(Y3(X), n+1)

# Compute the expectation E[Y3]
E_Y3 = np.mean(Y3(X))

# Compute the scaled difference n^r · (Dn(Y) - E[Y])
r_values = [1/4, 1/3, 1/2]
scaled_diffs = []

for r in r_values:
    scaled_diff = np.arange(1, num_terms+1)**r * (partial_sums_3 - E_Y3)
    scaled_diffs.append(scaled_diff)

# Plot the evolution of n^r · (Dn(Y) - E[Y]) for different values of r
for r, scaled_diff in zip(r_values, scaled_diffs):
    plt.plot(range(1, num_terms+1), scaled_diff, label='r={}'.format(r))

plt.xlabel('Number of terms (n)')
plt.ylabel('Scaled Difference')
plt.title('Evolution of n^r · (Dn(Y) - E[Y]) for Y3')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the number of simulations and the values of n to consider
num_simulations = 1000
n_values = [1, 10, 100, 1000]

# Define the random variable
def Y4(X):
    return np.tan(np.pi * (X - 1/2))

# Function to compute the partial sum Dn(Y)
def partial_sum(Y, n):
    return np.sum(Y[:n])

# Function to compute the empirical distribution function
def empirical_distribution_function(samples, u):
    return np.mean(samples <= u)

# Generate samples from the uniform distribution U(0, 1)
X = np.random.uniform(0, 1, num_simulations)

# Perform simulations for different values of n
for n in n_values:
    # Compute the partial sums for the random variable Y4
    partial_sums_4 = np.zeros(num_simulations)

    for k in range(num_simulations):
        partial_sums_4[k] = partial_sum(Y4(X[:n]), n)

    # Plot the empirical distribution function or build a histogram
    u_values = np.linspace(-5, 5, 100)
    frequencies = np.zeros_like(u_values)

    for i, u in enumerate(u_values):
        frequencies[i] = empirical_distribution_function(partial_sums_4, u)

    plt.plot(u_values, frequencies, label='n={}'.format(n))

plt.xlabel('Value (u)')
plt.ylabel('Empirical Distribution Function')
plt.title('Empirical Distribution Function for D(k)n iid ∼ Dn(Y4)')
plt.legend()
plt.show()
```
